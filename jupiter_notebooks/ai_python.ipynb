# Three Exercises to Get Started

### *You're encouraged to test your own ideas as you're reading this section!*

## (1) Debug

Here's a sample program with a goal to allow the user to choose between French and Spanish and teach them how to say hello in that language. The program currently has two errors. You are likely able to spot them on your own at this point in the course, but it can be tricky to find errors in your own code (even when they're simple), especially if you've been staring at your computer screen for a while.

Consider choosing an AI chatbot and putting in a program that currently has a syntax or logic issue. If you're looking for an accurate response, be *incredibly specific* about what you're looking for.

**Find the Error(s)**  

```python
print("Hello! Which language would you like to learn?")

print("Choose:\n\t(1) French\n\t(2) Spanish")
users_choice = int(input("Select a number: "))

if users_choice == 1:
    print("HELLO in French is BONJOUR!")
    print("Bonjour!")
elif users_choice = 2:
    print("HELLO in Spanish is HOLA!")
    print("Hola!")
else
    print("Sorry, we do not teach that language yet.")
```

ChatGPT reviewed the program above and identified there are **two syntax errors** and provided explanations of what's wrong, what Python expects, a fix for each error, and why each skill matters. 

This example is simple and you can likely correct these two errors using debugging tools in your IDE. You don't need to reference AI for everything! Sometimes reading error messages or Googling them is all it takes. When you're looking for more details about the *why* or for additional conversation or practice to avoid this error in the future, AI chatbots are great.

## (2) Write

Proceed with caution for this step. If your goal is to *learn* a skill, you may take a significant part of the learning away from yourself by having AI design the entire program. However, AI can generate entire programs. The more specific you are, the better. 

**Sample prompt #1:** I recently learned basic skills regarding lists, loops, and functions in Python. I would like to learn more about what I can do with lists in a Python program. Build a program that incorporates a bunch of list skills. I'm also a big Buffalo Bills fan if you want to include Bills info in the lists!

**Sample prompt #2:** Can you show me a bunch of list skills in a Python program? I like to eat different fruits! I'd also like to learn how to analyze a data set of fruits by using list methods.

**Sample prompt #3:** I love to play the game Wordle. Can you use this text file potential_answers.txt as the list of words that can be potential answers and teach me how to build a game like Wordle using Python?

## (3) Learn

You can share ideas about what you've learned so far, what you're struggling with, or what you hope to learn next! If you aren't sure what you want to learn next, you can ask for ideas based on what you know so far!

**Sample prompt #1**: I'm learning Python and I was looking at resources about lists. I noticed tuples and dictionaries on a few sites too. Explain the differences between lists, tuples, and dictionaries. Explain it to me like I'm 12.

***Note***: I like using "Explain it to me like I'm 12" for skills that are brand new to me to learn some foundational background knowledge. (I do this for non-Python questions too.) Use at your own risk as that prompt can also oversimplify the response. I recommend changing the age from 12 to 17 to 25 to 45 to 70 (or whatever age(s) you want) - Try it out and compare your responses!

**Sample prompt #2**: I accidentally moved a file into the wrong folder. The file is currently in the folder ch05_loops and I'd like to move the file into a folder called ch06_lists. Can you show me how to move the file to the correct folder using the terminal? The file's name is gratitude_list.py. After that, can you teach me some other important command line prompts?

**Sample prompt #3**: Here are some things I know how to do/create in Python: input, if statements, for loops, while loops, lists, functions. Can you make a 5-week plan for me to learn 10 more skills (2 per week)? 

***Note***: Use prompts like #3 when you're interested in a longer plan. You can (and should) reply with any skills in the response that you're already comfortable with that you forgot to mention in your initial prompt.

## Hallucinations - CAUTION!



After typing all of the info above, I pasted all of the text into a Gemini chat and asked for strengths and areas of improvement for this entire file. I did not make changes to the above info as I wanted you to see this AI-generated response separately from what I originally wrote. The #1 recommended change was to tell you about hallucinations and the importance of verification. Here's a snippet of my conversation with Gemini on October 22, 2025.

Gemini: *Although the tutorial mentions not needing AI for everything, it could more strongly warn students about the risk of "hallucination" (AI generating confidently incorrect code or facts). A line urging students to always test generated code and verify facts would be beneficial.*

Paige: *What are the top two ways to avoid hallucinations?*

Gemini: *The top two ways to avoid or mitigate hallucinations in large language models (LLMs) like AI chatbots are:*

***Grounding the Response with Retrieval-Augmented Generation (RAG):*** *This involves providing the AI with verified, external information (like official documentation, specific data, or a set of trusted documents) at the time of the query. The AI then uses this provided context as its primary source of truth, making it less likely to "make up" facts. Essentially, the AI is asked to answer only based on the facts you give it.* üìö


***Explicitly Asking the AI to Cite its Sources and Verify:*** *You can prompt the model to confirm the information and, if possible, provide direct citations or links to where it found the data.
